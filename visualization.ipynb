```python
# Read the CSV file into a DataFrame
df = pd.read_csv("combined.csv")

# Define the markers and colors for each newspaper
markers = {'Guardian': 'o', 'Times': 's', 'Sun': '^', 'Mail': 'd'}
colors = {'Guardian': 'blue', 'Times': 'red', 'Sun': 'green', 'Mail': 'orange'}

# Plot the positions of the words for each newspaper
fig, ax = plt.subplots(figsize=(6, 6))
for newspaper in markers.keys():
    newspaper_df = df[df['Newspaper'] == newspaper]
    plt.scatter(newspaper_df['Position'], newspaper_df['filename'], marker=markers[newspaper], color=colors[newspaper], label=newspaper)

# Set the x-axis and y-axis labels
plt.xlabel("Position")
plt.ylabel("Words")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
